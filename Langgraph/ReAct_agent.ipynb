## **Custom Tool**

```python
from langchain.tools import tool
@tool
def personal_information(name:str):

    """
    This is a tool to get the personal information of people.
    """
    info = {
        "Atul Rai":{
            "name":"Atul Rai",
            "age":28,
            "location":"India",
            "education":"B.Tech",
            "occupation":"Software Engineer"
        },
        "Alex":{
            "name":"Alex",
            "age":25,
            "location":"Ney York",
            "education":"B.Tech",
            "occupation":"Software Engineer"
        }
    }

    return info.get(name, "person not found")

```

```python
personal_information.invoke("Alex")
```

## Always remember the best way to use tool is to create a wrapper function and decorate it with @tool.

```python
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper

@tool
def wikipedia_search(query:str):
    """
    This is a tool to search wikipedia for a given query.
    """
    wikipedia_wrapper = WikipediaQueryRun(api_wrapper = WikipediaAPIWrapper())
    return wikipedia_wrapper.invoke(query)

wikipedia_search.invoke("Reinforcement learning in computer vision")
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

@tool
def duck_search(query:str):
    """
    This is a tool to search duckduckgo for a given query.
    """
    duck_search = DuckDuckGoSearchRun()
    return duck_search.invoke(query)

duck_search.invoke("What is the current price of gold in INR ?")
```

## **Tool Binding**

```python
tools = [personal_information, wikipedia_search, duck_search]
llm_with_tools = llm.bind_tools(tools)
```

```python
response = llm_with_tools.invoke("What are trend in AI ?")
print(response.content)
print(response.tool_calls)
```

## **LangGraph Creation**

```python
from typing import TypedDict, List
from langchain_core.prompts import ChatPromptTemplate

class graph_schema(TypedDict):
    messages:List
```

```python
def llm_node(state:graph_schema) -> graph_schema:

    messages = state["messages"]

    prompt = ChatPromptTemplate.from_messages(
        [

            ("system", "You are a helpful assistant that can answer the questions using the tools provided to you."),
            ("human","{input}")
        ]
    )

    chain = prompt | llm_with_tools

    response = chain.invoke({"input":messages})
    
    state["messages"] = messages + [response]

    return state
```

```python
# from langgraph.prebuilt import ToolNode  This can be used to create a tool node and use it in the graph. But we will create tool node manually.
from langchain_core.tools import tool
from langchain_core.messages import ToolMessage, HumanMessage
def tool_node(state:graph_schema)->graph_schema:
    messages = state["messages"]

    tools_by_name = {tool.name: tool for tool in tools}

    tool_result = []

    for tool_call in messages[-1].tool_calls:
        tool = tools_by_name[tool_call["name"]]

        observation = tool.invoke(tool_call["args"])

        tool_result.append({ToolMessage(name=tool.name, content=observation, tool_call_id=tool_call["id"])})

    state["messages"] = messages + tool_result

    return state
```

## **Create conditional edge function**

```python
def if_tool_call(state:graph_schema)->str:

    last_message = state["messages"][-1]

    if last_message.tool_calls:
        return "tool_node"
    else:
        return "end"
    
```

## **Create state graph**

```python
from langgraph.graph import StateGraph, START, END

graph = StateGraph(graph_schema)

graph.add_node("llm_node", llm_node)
graph.add_node("tool_node", tool_node)

graph.add_edge(START, "llm_node")
graph.add_conditional_edges("llm_node", if_tool_call, {"tool_node": "tool_node", "end": END})
graph.add_edge("tool_node", "llm_node")
graph.add_edge("llm_node", END)

react_graph = graph.compile()

from IPython.display import Image, display

Image(react_graph.get_graph().draw_mermaid_png())
```
