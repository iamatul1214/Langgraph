## **Prompts in langgraph**

```python
from dotenv import load_dotenv
import os
load_dotenv()

if os.getenv("GROQ_API_KEY") is None:
    raise ValueError("GROQ_API_KEY is not set")
else:
    print("GROQ_API_KEY is set")
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model_name="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY"),
)
llm
```

```python
from langchain_core.prompts import ChatPromptTemplate

my_prompt = ChatPromptTemplate.from_messages(

    [
        ("system","You are a helpful linkedin post writer. You are given a topic and you need to write a post about it."),
        ("human", "I want to create a post about {topic}.")
    ]
)
```

```python
prompt_filled = my_prompt.invoke({"topic":"The importance of data privacy in the digital Age"})
```

```python
llm.invoke(prompt_filled)
```

## **Chain in prompts**

#### We know that we have an LLM object and also a my_prompt object. Now, let's create a chain using langchain between both so that they can work in series.

```python
chain = my_prompt | llm
output = chain.invoke({"topic":"The importance of data privacy in the digital Age"})
output
```

```python
output.content
```

```python
chain = my_prompt | llm
output = chain.invoke({"topic":"The use of AI in the movies"})
output
```
